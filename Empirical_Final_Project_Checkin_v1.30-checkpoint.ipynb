#### Categorizing the weather condition  into 7 varialbes

```python
def categorize_weather(weather):
    if 'Clear' in weather:
        return 'Clear'
    elif 'Cloud' in weather or 'Overcast' in weather:
        return 'Cloudy'
    elif 'Rain' in weather or 'Drizzle' in weather or 'Thunderstorm' in weather:
        return 'Rainy'
    elif 'Fog' in weather or 'Haze' in weather or 'Mist' in weather or 'Smoke' in weather or 'Dust' in weather or 'Sand' in weather or 'Ash' in weather or 'Squalls' in weather:
        return 'Foggy'
    elif 'Snow' in weather or 'Sleet' in weather or 'Freezing' in weather:
        return 'Snowy'
    else:
        return 'Other'

df_1['Weather_Category'] = df_1['Weather_Condition'].apply(categorize_weather)

df_1['Weather_Category'].unique()
```

 # Using ML for predict if accident may take today on any of the Interstate with certain temprature & weather conditions then what can be the severity level for that Accident 

```python
from sklearn.preprocessing import LabelEncoder
data=maindata.copy(deep=True)

# Sample a smaller subset of data
data = data.sample(frac=0.1) # sample 10% of the data

# Select relevant features for prediction
features = ['Temperature(F)', 'Humidity(%)', 'Pressure(in)', 'Visibility(mi)', 'Wind_Speed(mph)', 'Traffic_Signal']
le = LabelEncoder()
data['Street'] = le.fit_transform(data['Street'])
# Split the data into training and testing sets
train_data = data[data['Start_Time'] < '2023-04-25'][features]
train_target = data[data['Start_Time'] < '2023-04-25']['Severity']

# Handle missing values with mean imputation
imputer = SimpleImputer(strategy='mean')
train_data = imputer.fit_transform(train_data)

# Train the model
rf_model = RandomForestClassifier(n_estimators=100, random_state=0)
rf_model.fit(train_data, train_target)

# Make predictions for one day
date = datetime.date.today()

prediction_data = pd.DataFrame(columns=features)

for i in range(24):
    prediction_data = prediction_data.append({'Temperature(F)': 10.1, 'Humidity(%)': 19.0, 'Pressure(in)': 19.62, 
                                               'Visibility(mi)': 8.0, 'Wind_Speed(mph)': 10.1, 'Traffic_Signal': True}, 
                                               ignore_index=True)

prediction_data = imputer.transform(prediction_data)
predictions = rf_model.predict(prediction_data)

# Calculate the severity score for the day
severity_score = sum(predictions) / len(predictions)

print('Predicted severity score for', date, ':', severity_score)


```

```python

train_accuracy = rf_model.score(train_data, train_target)


print('Training accuracy:', train_accuracy)
      
```

The most of the accidents are occuring on the rainy day. The wind direction is West-Northwest. But the Important point is that serverity level 4 has high impact on the road when compared to other serverity Levels.

```python
from sklearn.cluster import DBSCAN
from sklearn import metrics
import numpy as np

# define the epsilon radius for DBSCAN
eps = 0.1 / 6371.0  # 200 feet in kilometers
df_fil=maindata.sample(frac=0.1)

# filter data by accident severity > 2
df_filtered = df_fil[df_fil['Severity'] > 2]

# convert latitude and longitude to radians
df_filtered['Start_Lat_Rad'] = np.radians(df_filtered['Start_Lat'])
df_filtered['Start_Lng_Rad'] = np.radians(df_filtered['Start_Lng'])

# compute pairwise distances between accidents using haversine formula
from sklearn.neighbors import DistanceMetric
dist = DistanceMetric.get_metric('haversine')
X = df_filtered[['Start_Lat_Rad', 'Start_Lng_Rad']]
D = dist.pairwise(X)

# run DBSCAN on the pairwise distances
db = DBSCAN(eps=eps, min_samples=5, metric='precomputed').fit(D)
labels = db.labels_

# print the number of clusters
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
print('Estimated number of clusters: %d' % n_clusters_)

# plot the clusters on a scatter plot
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(10,10))
for i in range(-1, n_clusters_):
    subset = df_filtered[labels == i]
    if i == -1: # not in the range
        col = 'k'
        m = 'x'
    else: # in the range on the hotspot
        col = np.random.rand(3,)
        m = 'o'
    plt.scatter(subset['Start_Lng'], subset['Start_Lat'], s=20, marker=m, color=col)

plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Accident Clusters (Severity > 2)')
plt.show()
```

```python
# Subset the data to only include the relevant road features
road_features = df_filtered[['Amenity', 'Bump', 'Crossing', 'Give_Way', 'Junction', 'No_Exit', 'Railway', 'Roundabout', 'Station', 'Stop',  'Traffic_Signal', 'Turning_Loop', 'Severity']]

# Group the data by each road feature and count the number of accidents in each severity level
feature_counts = road_features.groupby(['Severity']).sum()

# Normalize the counts to get proportions instead of raw counts
normalized_counts = feature_counts.div(feature_counts.sum(axis=1), axis=0)

# Plot the normalized counts for each road feature
plt.figure(figsize=(12,6))
sns.barplot(data=normalized_counts, orient='h')
plt.title('Proportion of Accidents by Severity for each Road Feature based on the Hotspot Identified')
plt.xlabel('Proportion') # percentage of the accidents at Particular Feature
plt.ylabel('Road Feature')
plt.show()
```

```python
# Drop rows with missing 'Street' values
accidents = df_4.dropna(subset=['Street'])

# Filter the data to include only accidents that happened on interstate highways
interstate_accidents = accidents[accidents['Street'].str.contains('I-')]
```

```python
# Group the data by state and count the number of accidents
state_accidents = interstate_accidents.groupby('State').agg({'ID': 'count'}).rename(columns={'ID': 'Accidents'})

# Sort the data in descending order of accidents
state_accidents = state_accidents.sort_values(by='Accidents', ascending=False)

# Print the top 10 states with the highest number of accidents
print(state_accidents.head(10))
```

```python
# Create a pie chart of the top 10 states with the highest number of accidents
plt.figure(figsize=(8, 8))
plt.pie(state_accidents.head(10)['Accidents'], labels=state_accidents.head(10).index, autopct='%1.2f%%')
plt.title('Top 10 States with the Highest Number of Accidents')
plt.show()
```

```python
# Group the data by state and interstate highway and count the number of accidents
state_interstate_accidents = interstate_accidents.groupby(['State', 'Street']).agg({'ID': 'count'}).rename(columns={'ID': 'Accidents'})

# Sort the data in descending order of accidents
state_interstate_accidents = state_interstate_accidents.sort_values(by='Accidents', ascending=False)

# Print the top interstate highways for each state
for state in state_accidents.index[:10]:
    print(f'Top interstate highways in {state}:')
    print(state_interstate_accidents.xs(state).head(5))
    print()
```
